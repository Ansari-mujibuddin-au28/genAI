```python
### wite the content read the content ==== seek()

def count_line(file):
    with open(file,'r') as f:
        lines=f.readlines()
        total_lines=len(lines)
        total_words=sum(len(line.split())for line in lines)
        total_chat=sum(len(line) for line in lines)
    return total_lines,total_words,total_chat
file_path='total.txt'
total_lines,total_words,total_chat = count_line(file_path)
print(f"total lines{total_lines},total words{total_words}, total chat {total_chat}")
```
