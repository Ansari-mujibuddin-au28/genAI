```python
### loop control statemnets are being used to termianate a loop where the condition is met of if we want to have a control a loop
### break ---> to termiante the loop when a certain condition is met

for i in range(51):
    if i==18:
        break
    else:
        print(i)
```

```python
### loop control statemnets are being used to termianate a loop where the condition is met of if we want to have a control a loop
### continue --> it is used to skip the value if the condition is met
 
for i in range(51):
    if i%2==0:
        print(i)
    else:
        continue
```
