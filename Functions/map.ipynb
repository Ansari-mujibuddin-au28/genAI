###Map()

### map can itterate in a list and returns a map object(AN iterator)

### map fn is particulary useful for transforming the data or applying an operation on ALL items of the iterable
### Map() is going to help out avoiding unncesseary lines of code we can apply the operation on all the value or all items in an iterable and this is the sign of a good profremmer

### syntax for map()
## list(map(funnitterable))

```python
### squaring all the items of a list

def squaring(list_item):
    empty_list=[]
    for item in list_item:
        suared_item=item**2
        empty_list.append(suared_item)
    print(empty_list)
```

```python
squaring([1,2,3,4])
```

```python
### simpler problem way using lamda and map
print(list(map(lambda x:x**2,[1,2,3,4])))
```

```python
### Map function with basic function in it 
def sqares_num(num):
    return num**2
print(list(map(sqares_num,[1,2,3,4])))
```

```python
### map function with multiple list or itterable
### [1,2,3,4] <---> [5,6,7,8] ---> addition ---> [6,8,11,6]

num_1 = [1,4,5,8]
num_2 = [8,5,6,7]

print(list(map(lambda x,y:x+y,num_1,num_2)))
```

```python
### map fn to str value to int value


print(list(map(int,['1','2','3','4'])))
```

```python
emp_name=['mujib',"abc","dfg",'hjk','zxc']
print(list(map(str.upper,emp_name)))
```

```python
### retrive the value inside a dic using map()
emp_name=[{'name':"mujib","age":20,'qualification':"btech"}
          ,{'name':"mujib123","age":30,'qualification':"dip"},{'name':"mujib456","age":60,'qualification':"inter"}]

def age(list_item):
    return list_item['age']
print(list(map(age,emp_name)))
```
