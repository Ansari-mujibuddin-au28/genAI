### Filter Function

### it is going to retrevive on those items which are meeting a certain condition in an iterable


### Filter function consruction an iterator from element of an iterable for which a conditin remains true
### syntex for filter 

List(filter(fun,itterable))

```python
### to retrevie the only those value which are even
numbers = [1, 2, 3, 4, 5, 6, 7]

def even(value):
    return value % 2 == 0

print(list(filter(even, numbers)))
```

```python
### solve with lamda
```

```python
print(list(filter(lambda x:x%2==0,numbers)))
```

```python
### greater the 5
print(list(filter(lambda x:x>5,numbers)))
```

```python
numbers = [1, 2, 3, 4, 5, 6, 7.948,666,66,22,88]
print(list(filter(lambda x:x>5 and x%2==0,numbers)))
```
