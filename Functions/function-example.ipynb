"""1 deg cecius = 9/5 +32
45 deg c = 45*9/5 +32 F

45 deg F =(45-32) = *5/9 deg c """

```python
### define a function that will behave as password strength checker

""" strong or week passwoed -->  """
def passwordfn(password):
    special_chars = "!@#$%^&*()_-"

    if len(password) < 8:
        print(f"The entered password '{password}' is weak")

    elif not any(char.islower() for char in password):
        print(f"The entered password '{password}' is weak")

    elif not any(char.isupper() for char in password):
        print(f"The entered password '{password}' is weak")

    elif not any(char.isdigit() for char in password):
        print(f"The entered password '{password}' is weak")

    elif not any(char in special_chars for char in password):
        print(f"The entered password '{password}' is weak")

    else:
        print("This is a strong password ✅")
```

```python
passwordfn("Test123")      # weak (no special char)
passwordfn("Test@123")     # strong
```

```python
### check wheater if the current string is a palindrome or not
### palindrom str is going to have similar from both the side if reverse from right to left and left to right
## ---> aaabbaa 
## if it not same then is  not palindrome
### reverse string ::-1

def palindrome(string):
    string_value=string.lower().replace(" ","")
    if string_value == string_value[::-1]:
        print(f"valid palindrome")
    else:
        print("not valid palindrome")
```

```python
palindrome("aabb  aa")
```

```python
### Calculate the total cost of items in a shopping cart

cart = [{'item1':"apple",'cost':5,"qantity":20},{'item2':"banana",'cost':7,"qantity":40},{'item3':"avacado",'cost':8,"qantity":90}]

def cart_total(cart):
    total_cost =0
    for item in cart:
        total_cost+=item["cost"]*item['qantity']
    print("the total cost",total_cost)
```
